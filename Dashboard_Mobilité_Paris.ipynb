## Prétraitement des données

```python
# Utiliser pd.melt pour transformer les années en une seule colonne
df = pd.melt(df, id_vars=['Localisation', 'Date_instal', 'l_ar', 'Code_Lat', 'Code_Long'], var_name='year', value_name='comptages_annuels')

# Convertir les colonnes aux types appropriés
df['Localisation'] = df['Localisation'].astype(str)
df['year'] = df['year'].astype(int)
df['comptages_annuels'] = df['comptages_annuels'].astype(int)

print(df)
```

```python
# Enregistrement de la data Base CSV/ année
df.to_csv('Velo_Paris_new.csv', index=False)
```

```python
#Trier par le nombre de taxi par année (décroissant)
selected_year = 2023
df_selected_year = df[df.year == selected_year]
df_selected_year
```

```python
# Trier par année
df_selected_year_année = df_selected_year.sort_values(by="comptages_annuels", ascending=False)
df_selected_year_année
```

```python
# Calculer la différence de nombre de velo entre l'année sélectionnée et l'année précédente
def calculate_comptages_annuels(input_df, input_year):
  selected_year_data = input_df[input_df['year'] == input_year].reset_index()
  previous_year_data = input_df[input_df['year'] == input_year - 1].reset_index()
  selected_year_data['différence_comptage'] = selected_year_data.comptages_annuels.sub(previous_year_data.comptages_annuels, fill_value=0)
  selected_year_data['différence_comptage_absolue'] = abs(selected_year_data['différence_comptage'])
  return pd.concat([selected_year_data.Localisation, selected_year_data.l_ar, selected_year_data.comptages_annuels, selected_year_data.différence_comptage, selected_year_data.différence_comptage_absolue], axis=1).sort_values(by="différence_comptage", ascending=False)

df_différence_comptage_sorted = calculate_comptages_annuels(df, selected_year)
df_différence_comptage_sorted
```

```python
# Filtrer le nombre de velo avec une différence > 1000000
df_greater_1000000 = df_différence_comptage_sorted[df_différence_comptage_sorted.différence_comptage_absolue > 1000000]
df_greater_1000000
```

```python
# % Des compages sup à > 1000000
int((len(df_greater_1000000)/df_différence_comptage_sorted.Localisation.nunique())*100)
```

#### Choroplèthe avec plotly.express

```python
import pandas as pd
import geopandas as gpd
import plotly.express as px
import plotly.graph_objects as go

# Charger les données GeoJSON des arrondissements de Paris
Paris_geojson = 'C:/Users/kaize/OneDrive/Bureau/Appli/arrondissements.geojson'
gdf = gpd.read_file(Paris_geojson)

# Charger le fichier CSV contenant les comptages annuels 
Velo_CSV = 'C:/Users/kaize/OneDrive/Bureau/Appli/Velo_Paris_new.csv'
Paris_velo = pd.read_csv(Velo_CSV, sep=',')  # Changez le séparateur

# Assurez-vous que les colonnes correspondent pour la jointure
Paris_velo['l_ar'] = Paris_velo['l_ar'].astype(str)
gdf['l_ar'] = gdf['l_ar'].astype(str)

# Reprojeter les géométries (EPSG:4326 pour lat/lon)
gdf = gdf.to_crs(epsg=4326)

# Calculer les centroides des arrondissements reprojetés
gdf['centroid'] = gdf.geometry.centroid
gdf['Code_Lat'] = gdf['centroid'].y
gdf['Code_Long'] = gdf['centroid'].x

# Fusionner les données géographiques avec les comptages annuels
gdf = gdf.merge(Paris_velo, on='l_ar', how='left')

# Nettoyer les colonnes après la fusion pour éviter la confusion
gdf = gdf.drop(columns=['Code_Lat_y', 'Code_Long_y'], errors='ignore')  # Supprimer les colonnes inutiles
gdf = gdf.rename(columns={'Code_Lat_x': 'Code_Lat', 'Code_Long_x': 'Code_Long'}, errors='ignore') 

# Créer une carte avec des points de comptages annuels
fig = px.scatter_mapbox(
    gdf,
    lat='Code_Lat',
    lon='Code_Long',
    color='comptages_annuels',
    size='comptages_annuels',
    color_continuous_scale=px.colors.cyclical.IceFire,
    size_max=40,
    zoom=11,
    center={"lat": 48.8566, "lon": 2.3522},
    mapbox_style="open-street-map",
    hover_name='l_ar',
    hover_data=['comptages_annuels']
)

# Ajouter le contour de Paris
contour_Paris_geojson = 'C:/Users/kaize/OneDrive/Bureau/Appli/Contours.geojson'
contour_gdf = gpd.read_file(contour_Paris_geojson)
contour_Paris_geojson = contour_gdf.geometry.__geo_interface__

for feature in contour_Paris_geojson['features']:
    coords = feature['geometry']['coordinates'][0]
    lon = [coord[0] for coord in coords]
    lat = [coord[1] for coord in coords]
    fig.add_trace(go.Scattermapbox(
        lon=lon,
        lat=lat,
        mode='lines',
        line=dict(width=2, color='black'),
        showlegend=False
    ))

# Mettre à jour la mise en page de la carte
fig.update_layout(
    template='plotly_dark',
    plot_bgcolor='rgba(0, 0, 0, 0)',
    paper_bgcolor='rgba(0, 0, 0, 0)',
    margin=dict(l=0, r=0, t=0, b=0),
    height=800
)

# Afficher la carte
fig.show()
```

### Graphique treemap

```python
import pandas as pd
import plotly.express as px
import geopandas as gpd

# Charger les données GeoJSON
geojson_path = 'C:/Users/kaize/OneDrive/Bureau/Appli/arrondissements.geojson'
gdf = gpd.read_file(geojson_path)

# Charger les données CSV
Paris_velo = 'C:/Users/kaize/OneDrive/Bureau/Appli/Velo_Paris_new.csv'
df = pd.read_csv(Paris_velo)

# Assurez-vous que les colonnes correspondent pour la jointure
df['l_ar'] = df['l_ar'].astype(str)
gdf['l_ar'] = gdf['l_ar'].astype(str)

# Fusionner les données géographiques avec les comptages annuels
gdf = gdf.merge(df, on='l_ar', how='left')

# Préparer les données pour le treemap
# Nous allons utiliser 'Localisation' comme niveau principal et 'Date_instal' comme sous-niveau
# Nous allons utiliser 'comptages_annuels' pour la taille des rectangles et la couleur

# Assurez-vous que les colonnes sont présentes dans le DataFrame
assert 'Localisation' in gdf.columns
assert 'Date_instal' in gdf.columns
assert 'comptages_annuels' in gdf.columns

# Créer le treemap
fig = px.treemap(
    gdf, 
    path=['Localisation', 'Date_instal'], 
    values='comptages_annuels',
    color='comptages_annuels',  # Utilisez 'comptages_annuels' pour la couleur
    color_continuous_scale='viridis',  # Vous pouvez ajuster la palette de couleurs ici
    hover_data={'Localisation': True, 'comptages_annuels': True, 'Date_instal': True}
)

# Afficher le treemap
fig.show()
```
